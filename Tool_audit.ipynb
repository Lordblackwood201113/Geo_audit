```python
#Construire le Toggle principal et les autres Toggle

toggle_princ = widgets.ToggleButton(
    tooltip = "Menu",
    icon = "bars",
    layout=widgets.Layout(width="28px", height="28px", padding=padding) ,
)

file = widgets.Button(
    icon = "folder-open",
    tooltip = "Ouvrir un fichier",
    button_style="primary",
    layout=widgets.Layout(width="28px", height="28px", padding= "0px 0px 0px 0px") ,
)

processing = widgets.Button(
    icon = "cogs",
    tooltip = "think",
    button_style="primary",
    layout=widgets.Layout(width="28px", height="28px", padding= "0px 0px 0px 0px") ,
)

database = widgets.Button(
    icon = "database",
    tooltip = "Table attributaire",
    button_style="primary",
    layout=widgets.Layout(width="28px", height="28px", padding= "0px 0px 0px 0px") ,
)

analyse = widgets.Button(
    icon = "terminal",
    tooltip = "Analyse",
    button_style="primary",
    layout=widgets.Layout(width="28px", height="28px", padding= "0px 0px 0px 0px") ,
)

download = widgets.Button(
    icon = "download",
    tooltip = "Téléchargement",
    button_style="primary",
    layout=widgets.Layout(width="28px", height="28px", padding= "0px 0px 0px 0px") ,
)
#grid_template_columns="repeat(2, 40px)", width="80px")
bloc_princi = widgets.GridBox([file, processing, database, analyse, download], layout=widgets.Layout(grid_gap="0px"))

param = widgets.VBox([toggle_princ])
```

```python
#Evènement des boutons

def toggle_princ_event (change) :
    if change["new"]:
          # Afficher le menu complet
          param.children = [toggle_princ, bloc_princi]
    else:
        param.children = [toggle_princ]
        
toggle_princ.observe(toggle_princ_event, "value")

#--------------------------------------------------------------------------------

#Evènement du widget de téléchargement (dwd_ctrl)

def download_event(b) :
    name_layer = []
    for layer in m.layers:
        if isinstance(layer, GeoData):
            name_layer.append(layer.name)
    dropdown_dwd.options = name_layer
    m.add(dwd_ctrl)
    download.disabled = True
    
download.on_click(download_event)

def close_dwd_event(b) :
    m.remove_control(dwd_ctrl)
    download.disabled = False
    
close_dwd.on_click(close_dwd_event)
    
def dropdown_dwd_event(change) :
    attributes = []
    if change["new"]:
        dwd_layer = get_layer_by_name(m, dropdown_dwd.value)
        
        feature_layer = dwd_layer.data['features']
        
        for feature in feature_layer:
            props = feature['properties']
            attributes.append(props)
        
        layer_csv = pd.DataFrame(attributes)

        csv_buffer = StringIO()
        layer_csv.to_csv(csv_buffer, index=False)
        csv_data = csv_buffer.getvalue()

        filename = '{}.csv'.format(dropdown_dwd.value)
        b64 = base64.b64encode(csv_data.encode())
        payload = b64.decode()
        
        html_button.value = html_buttons.format(payload=payload, filename=filename)
        
dropdown_dwd.observe(dropdown_dwd_event, "value")
    
    


#--------------------------------------------------------------------------------


#Widgets File_ctrl
def file_event(b) :
    Box_fc.children = [Header_fc, fc, text_fc, col_fc, Footer_fc]
    m.add(file_ctrl)
    file.disabled = True
file.on_click(file_event)

def close_fc_event(b) :
    fc.reset()
    m.remove_control(file_ctrl)
    file.disabled = False

    
close_fc.on_click(close_fc_event)

def reset_fc_event(b) :
    fc.reset()
    Box_fc.children = [Header_fc, fc, text_fc, col_fc, Footer_fc]
    
reset_fc.on_click(reset_fc_event)

def apply_fc_event(b) :
    Load_fc.style= {'bar_color': '#ffff00'}
    Load_fc.description = "Chargement : "
    Box_fc.children = [Header_fc, fc, text_fc, col_fc, Load_fc, Footer_fc]
    Load_fc.value = 2.0

    gdf = gpd.read_file(fc.selected, driver='LIBKML')
    
    Load_fc.value = 5.0
    
    Load_fc.value = 6.0

    geo_data = GeoData(geo_dataframe = gdf, name = text_fc.value,
    style={'fillColor': col_fc.value, 'opacity':0.05, 'weight':1.9, 'dashArray':'2', 'fillOpacity':0.6}, 
    hover_style={ "fillColor": 'blue', "fillOpacity": 0.5},
    )
    
    Load_fc.value = 8.0
    
    text_fc.value = ''
    col_fc.value = "#FFFFFF"
    m.add(geo_data)
    
    Load_fc.value = 10.0
    
    Load_fc.style=style={'bar_color': 'green'}
    Load_fc.description = "Terminer : " 
    """
    data = fc.selected
    
    gdf = gpd.read_file(data, driver='KML')
    geo_data = GeoData(geo_dataframe = gdf,
                   style={'color': 'black', 'fillColor': '#3366cc', 'opacity':0.05, 'weight':1.9, 'dashArray':'2', 'fillOpacity':0.6},
                   hover_style={'fillColor': 'red' , 'fillOpacity': 0.2},
                   name = 'Plantation')
    m.add(geo_data)
    """
    fc.reset()

apply_fc.on_click(apply_fc_event)

 
#html = HTML("""Hover over a state""")
#html.layout.margin = "0px 20px 20px 20px"
#control = WidgetControl(widget=html, position="bottomright")
#m.add(control)

#Widget Analyse spatiale

def as_event(b) :
    name_layer = []
    for layer in m.layers:
        if isinstance(layer, GeoData):
            name_layer.append(layer.name)
    dropdown_as_1.options = name_layer
    dropdown_as_2.options = name_layer
    
    m.add(as_ctrl)
    processing.disabled = True
    
processing.on_click(as_event)

def close_as_event(b) :
    m.remove_control(as_ctrl)
    processing.disabled = False

    
close_as.on_click(close_as_event)

def actual_as_event(b) :
    actual_as.disabled = True
    Box_as.children = [Header_as, Label_as_1, dropdown_as_1, Label_as_2, dropdown_as_2, Footer_as]
    
    name_layer = []
    for layer in m.layers:
        if isinstance(layer, GeoData):
            name_layer.append(layer.name)
    dropdown_as_1.options = name_layer
    dropdown_as_2.options = name_layer
    actual_as.disabled = False
    
    
actual_as.on_click(actual_as_event)

# Apply Analyse spatiale
def get_layer_by_name(map_obj, layer_name):
    for layer in map_obj.layers:
        if hasattr(layer, 'name') and layer.name == layer_name:
            return layer
    return None

def apply_as_event(b):
    # Récupérer les couches sélectionnées
   
    Box_as.children = [Header_as, Label_as_1, dropdown_as_1, Label_as_2, dropdown_as_2, Load_as, Footer_as]
    gdf1_layer = get_layer_by_name(m, dropdown_as_1.value)
    gdf2_layer = get_layer_by_name(m, dropdown_as_2.value)
    Load_as.value = 5.0
    # Assurer que les couches sélectionnées sont de type GeoData
    if isinstance(gdf1_layer, GeoData) and isinstance(gdf2_layer, GeoData):
        gdf1 = gdf1_layer.geo_dataframe  # Récupérer le GeoDataFrame de la couche 1
        gdf2 = gdf2_layer.geo_dataframe  # Récupérer le GeoDataFrame de la couche 2
        
        # Effectuer l'analyse spatiale (chevauchement)
        chevauchement = gpd.sjoin(gdf1, gdf2, how="inner", predicate='intersects')
        Load_as.value = 9.0
        
        # Créer une nouvelle couche GeoData à partir du résultat de l'analyse
        geo_data = GeoData(
            geo_dataframe=chevauchement,
            name="Couche invalide",
            style={'fillColor': "black", 'opacity': 1, 'weight': 1.9, 'dashArray': '2', 'fillOpacity': 0.6},
            hover_style={"fillColor": 'Yellow', "fillOpacity": 0.9}
        )
        
        # Ajouter la nouvelle couche à la carte
        m.add_layer(geo_data)
        Load_as.value = 10.0
        Load_as.style=style={'bar_color': 'green'}
        Load_as.description = "Terminer : "
    

apply_as.on_click(apply_as_event)







    
    
```
